```python
import random

# ASCII art for hangman stages
HANGMAN_PICS = [
    """
       ------
       |    |
            |
            |
            |
            |
    =========
    """,
    """
       ------
       |    |
       O    |
            |
            |
            |
    =========
    """,
    """
       ------
       |    |
       O    |
       |    |
            |
            |
    =========
    """,
    """
       ------
       |    |
       O    |
      /|    |
            |
            |
    =========
    """,
    """
       ------
       |    |
       O    |
      /|\\   |
            |
            |
    =========
    """,
    """
       ------
       |    |
       O    |
      /|\\   |
      /     |
            |
    =========
    """,
    """
       ------
       |    |
       O    |
      /|\\   |
      / \\   |
            |
    =========
    """
]

def choose_word():
    words = ['python', 'robotics', 'automation', 'programming', 'machine', 'learning']
    return random.choice(words)

def display_progress(word, guessed_letters):
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter + " "
        else:
            display += "_ "
    return display.strip()

def hangman():
    word = choose_word()
    guessed_letters = []
    attempts = 6

    print("Welcome to Hangman!")
    print("Try to guess the word letter by letter.")
    
    while attempts > 0:
        print(HANGMAN_PICS[6 - attempts])  # Display the current hangman stage
        print("Word:", display_progress(word, guessed_letters))
        print("Attempts left:", attempts)
        print("Guessed letters:", " ".join(guessed_letters))

        guess = input("Enter a letter: ").lower()
        
        if guess in guessed_letters:
            print("You already guessed that letter. Try a different one.")
            continue
        elif guess in word:
            print("Good guess!")
            guessed_letters.append(guess)
        else:
            print("Wrong guess.")
            attempts -= 1
            guessed_letters.append(guess)

        if all(letter in guessed_letters for letter in word):
            print("\nCongratulations! You guessed the word:", word)
            break
    else:
        print(HANGMAN_PICS[6])  # Show final stage of hangman
        print("\nSorry, you're out of attempts. The word was:", word)

hangman()
```
